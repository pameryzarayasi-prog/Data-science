```python
#1. Estudiantes inscritos en dos cursos
estudiantes_python = {"Ana", "Luis", "María", "Carlos", "Sofía", "Diego"}
estudiantes_javascript = {"María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria"}
#Encuentra:
#- Los estudiantes inscritos en ambos cursos.
#- Los estudiantes que solo están en Python.
#- Los estudiantes que solo están en JavaScript.
#- Todos los estudiantes sin repetir nombres.
print("Ambos cursos",estudiantes_python & estudiantes_javascript)
print("solo Python",estudiantes_python - estudiantes_javascript)
print("solo JavaScript", estudiantes_javascript-estudiantes_python)
print("todos", estudiantes_javascript| estudiantes_python)

```

```python
#2. Comparar inventarios de dos almacenes
almacen_arequipa = {"laptop", "teclado", "mouse", "monitor", "impresora", "audifonos"}
almacen_lima = {"laptop", "mouse", "monitor", "tablet", "webcam", "parlantes"}
#Determina:
#- Los productos disponibles en ambos almacenes.
#- Los productos exclusivos del almacén de Arequipa.
#- Los productos exclusivos del almacén de Lima.
#- Todos los productos disponibles sin repetir.
#- Si ambos almacenes tienen exactamente el mismo inventario.
print("Productos disponibles en ambos almacenes", almacen_arequipa&almacen_lima)
print("Productos solo en Arequipa", almacen_arequipa - almacen_lima)
print("Productos solo en Lima", almacen_lima - almacen_arequipa)
print("Todos los productos", almacen_lima | almacen_arequipa)
print("Todos los productos son iguales en los almacenes: ", almacen_lima == almacen_arequipa)

```

```python
#3. Validar permisos de un usuario
permisos_usuario = {"leer", "editar", "comentar", "descargar"}
permisos_requeridos = {"leer", "editar", "eliminar", "publicar"}
#Crea un programa que muestre:
#- Si el usuario posee todos los permisos requeridos.
#- Los permisos que le faltan.
#- Los permisos adicionales que tiene.
#- La cantidad total de permisos diferentes.
if len(permisos_requeridos-permisos_usuario)==0:
    print("Permisos completos")
else:
    print("No tiene todos los permisos")
#print("Usuario tienes todos los permisos requeridos: ",permisos_requeridos-permisos_usuario)
print("Permisos faltantes: ",permisos_requeridos-permisos_usuario)
print("Permisos adicionales: ",permisos_usuario-permisos_requeridos)
print("Cantidad de Permisos faltantes: ", len(permisos_requeridos-permisos_usuario))
```

```python
#4. Números presentes en varias listas
numeros_a = {2, 4, 6, 8, 10, 12, 14}
numeros_b = {4, 6, 9, 10, 13, 14, 16}
numeros_c = {1, 4, 6, 10, 11, 14, 18}
#Encuentra:
#- Los números presentes en los tres conjuntos.
#- Los números presentes en numeros_a y numeros_b, pero no en numeros_c.
#- Los números que solo pertenecen a numeros_a.
#- Todos los números diferentes.
print("Números presentes en 3 listas", numeros_a&numeros_b&numeros_c)
print("Números presentes en numeros_a y numeros_b", (numeros_a|numeros_b)-numeros_c)
print("Números solo pertenecen a numeros_a", numeros_a-(numeros_c|numeros_b))
print("Números presentes en 3 listas", numeros_a^numeros_b|(numeros_c-(numeros_a|numeros_b)))
```

```python
#5. Comparar habilidades de candidatos
habilidades_candidato_1 = {"Python", "Git", "SQL", "Docker", "React"}
habilidades_candidato_2 = {"Python", "JavaScript", "React", "Node.js", "Git"}
habilidades_requeridas = {"Python", "Git", "SQL", "Docker"}
#Determina:
#- Las habilidades que comparten ambos candidatos.
#- Las habilidades exclusivas de cada candidato.
#- Qué habilidades requeridas le faltan a cada candidato.
#- Qué candidato cumple con más requisitos.
#- Todas las habilidades disponibles sin repetir.
print("habilidades que comparten ambos candidatos",habilidades_candidato_1&habilidades_candidato_2)
print("habilidades exclusivas cantidato 1",habilidades_candidato_1- habilidades_candidato_2)
print("habilidades exclusivas cantidato 2", habilidades_candidato_2-habilidades_candidato_1)
if len(habilidades_requeridas-habilidades_candidato_1)>len(habilidades_requeridas- habilidades_candidato_2):
    print("El candidato 2 cumple más requisitos")
else:
    print("El candidato 1 cumple más requisitos")
print("Todas las habilidades", habilidades_candidato_2|habilidades_candidato_1|habilidades_requeridas)
```
