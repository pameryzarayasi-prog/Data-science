```python
#1. Consolidación de inventarios
#Combina tres inventarios provenientes de distintas sucursales. Cuando un producto aparezca en varias
#sucursales, suma sus cantidades y calcula también en cuántas sucursales está disponible.
inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3}
inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5}
inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7}
#Requisitos:
#● Crea un diccionario final con las claves cantidad_total y sucursales.
#● No modifiques los diccionarios originales.
#● Ordena el resultado alfabéticamente por producto
inventario_total=dict(inventario_norte)

for llave, valor in inventario_sur.items():
    if llave not in inventario_total:
        inventario_total[llave]=0
    inventario_total[llave]+=valor

for llave, valor in inventario_centro.items():
    if llave not in inventario_total:
        inventario_total[llave]=0
    inventario_total[llave]+=valor
    print(inventario_total)
```

```python
llave_ordenada=sorted(inventario_total.keys())
inventario_ordenado={}
for llave in llave_ordenada:
   inventario_ordenado[llave]= inventario_total[llave]
print(inventario_total)
```

```python
#2. Ranking de estudiantes con desempate
#Calcula el promedio de cada estudiante y genera un ranking de mayor a menor. Si dos estudiantes tienen el
#mismo promedio, ordénalos alfabéticamente.
calificaciones = {
 "Ana": [17, 18, 16, 19],
 "Luis": [15, 16, 17, 16],
 "Carla": [18, 17, 19, 18],
 "Pedro": [16, 16, 16, 16],
 "Lucía": [15, 17, 16, 16]
}
#Requisitos:
#● Guarda promedio, nota_máxima y nota_mínima por estudiante.
#● El ranking final debe ser una lista de tuplas.
estadistica= {}
for estudiante, notas in calificaciones.items():
    print(estudiante, notas)
    estadistica[estudiante]={
        "minimo":min(notas),
        "maximo":max(notas),
        "promedio":sum(notas)/len(notas)
        }

for estudiante, estad in estadistica.items():
    print(estudiante)
    print(estad)

sorted(estadistica.items())
```

```python
#4. Análisis de frecuencia de palabras 
#Construye un diccionario de frecuencias ignorando mayúsculas, signos de puntuación y palabras vacías. 
#Práctica de Python - Diccionarios 
texto = "Python es poderoso; Python es flexible, y aprender Python puede ser muy divertido." 
palabras_vacias = {"es", "y", "ser", "muy", "puede"} 
#Requisitos:
 
#● Muestra las tres palabras más frecuentes. 
#● Resuelve el ejercicio sin usar collections.Counter.

contador={}
for palabra in texto.split(" "):
    palabra_limpia=palabra.lower().replace(";","").replace(",","").replace(".","")
    if palabra_limpia in palabras_vacias:
        continue

    if palabra_limpia not in contador:
        contador[palabra_limpia]= 0
    contador[palabra_limpia]+=1

    print(contador)
```

```python
#7. Sistema de votación con validación 
#Procesa una lista de votos y separa los votos válidos de los inválidos. 
candidatos = {"Ana", "Luis", "Carla"} 
votos = ["Ana", "Luis", "ana", "Carla", "Pedro", "Luis", "ANA", "", "Carla"] 
#Requisitos:
#● Los votos deben ser insensibles a mayúsculas. 
#● Registra conteo por candidato, votos inválidos y porcentaje obtenido. 
#● Determina si hay ganador o empate.

mapa={
    "ana":"Ana",
    "luis":"Luis",
    "carla":"Carla"
}
conteo={}

for voto in votos:
    voto_limpio=voto.lower()
    candidato=mapa.get(voto_limpio,"invalido")
    
    if candidato=="invalido":
         conteo.setdefault("Votos invalidos", 0)
         conteo["Votos invalidos"]+=1
         continue

    conteo.setdefault(candidato, 0)
    conteo[candidato]+=1
print(conteo)


total_votos=len(votos)
for candidato in candidatos:
     conteo[f"{candidato}_pct"]= conteo[candidato]/total_votos*100

conteo["votos invalidos pct"]=conteo["Votos invalidos"]/total_votos*100

print(conteo["votos invalidos pct"])
print(conteo[f"{candidato}_pct"])
```
