```python
#EJERCICIO 1. Buscar el mayor y el menor valor 
#Dada la tupla 
numeros = (15, 8, 42, 3, 27, 19, 5)
#obtén el número mayor, el menor y la diferencia entre ambos.
lista_max=max(numeros) 
lista_min=min(numeros)
resta=lista_max-lista_min
print(resta)
```

```python
#EJERCICIO 4. Estadísticas de una tupla Con 
notas = (18,15,20,11,14,17,19)
#calcula promedio, máxima, mínima y cantidad de aprobados. 

numero_notas=len(notas)
contador=0
for nota in notas:
    contador=contador + nota

Promedio=round(contador/numero_notas)
print(f"El Promedio es: {Promedio}")
print(max(notas))
print(min(notas))
```
