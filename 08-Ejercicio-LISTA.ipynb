```python
# EJERCICIO 1 Eliminar duplicados
numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4] 
numeros_unicos = []
for numero in numeros:
    if numero not in numeros_unicos:
        numeros_unicos.append(numero)

print(numeros_unicos)
```

```python
#EJERCICIO 2 Separar pares e impares 
#Dada una lista de números, crea dos listas: una con números pares y otra con números impares.
numeros = [12, 7, 9, 24, 31, 18, 5, 42, 13, 8]

numero_par = [
    numero
    for numero in numeros
    if numero % 2 ==0
]
numero_impar = [
    numero
    for numero in numeros
    if numero % 2 ==1
]
print(numero_par)
print(numero_impar)
```

```python
#EJERCICIO 3 Encontrar el segundo número mayor 
# #Dada una lista de números, encuentra el segundo valor más grande. 
numeros = [45, 12, 89, 34, 89, 23, 67, 10]

numero_unicos=[]
for numero in numeros:
    if numero not in numeros_unicos:
        numeros_unicos.append(numero)   

numeros_unicos.sort(reverse=True)      
print(numeros_unicos)
 
print(f"El numero segundo es: {numeros_unicos[1]}")


```

```python
#EJERCICIO 4 Contar palabras largas 
#Contar palabras largas, Dada una lista de palabras, cuenta cuántas palabras tienen más de 5 letras. 
palabras = ['python', 'sol', 'programacion', 'lista', 'computadora', 'if', 'variable'] 
contador= 0
palabras_largas=[]
for palabra in palabras:
    if len(palabra)>5:
     contador = contador + 1
     palabras_largas.append(palabra)
print(f"Las palabras con mas de 5 letras son:  {contador}")
print(f"Las palabras son:  {palabras_largas}")
```

```python
#EJERCICIO 5  Promedio de notas 
# Dada una lista de notas, calcula el promedio e indica si el estudiante aprobó o desaprobó. 
notas = [14, 18, 11, 16, 9, 13, 20] 
len(notas)
suma_notas=0
for nota in notas:
   suma_notas=suma_notas + nota
promedio_notas=round(suma_notas/len(notas),2)
print(promedio_notas)

if promedio_notas>10.5:
   print("El alumno esta aprobado")
else: 
   print("El alumno esta desaprobado")
```

```python
#EJERCICIO 8 Buscar elementos comunes
#Buscar elementos comunes 
#Dadas dos listas, encuentra los elementos que aparecen en ambas. 
lista1 = [1, 3, 5, 7, 9, 11] 
lista2 = [2, 3, 6, 7, 10, 11] 

lista_comun= []
for numero in lista1:
    if numero in lista2:
     lista_comun.append(numero)
print(f"Los numero en comun son los siguientes: [lista_comun]")
```

```python
#EJERCICIO 9 Buscar elementos únicos entre dos listas 
#Dadas dos listas, crea una nueva lista con los elementos que no se repiten entre ambas listas. 
lista1 = [4, 8, 12, 16, 20] 
lista2 = [8, 10, 12, 18, 20] 
lista_diferencia1=[
    numero
    for numero in lista2
    if numero not in lista1
]
lista_diferencia=[
    numero
    for numero in lista1
    if numero not in lista2
]
lista_diferente_completo=lista_diferencia + lista_diferencia1
print(lista_diferente_completo)
```

```python
#EJERCICIO 10. Invertir una lista sin usar reverse() 
#Dada una lista de colores, invierte la lista sin usar el método .reverse(). 
colores = ['rojo', 'azul', 'verde', 'amarillo', 'morado'] 

ultimo=len(colores)-1
al_reves=[]
while ultimo>=0:
  al_reves.append(colores[ultimo])
  ultimo=ultimo-1
print(al_reves)
```

```python
#EJERCICIO 11. Ordenar palabras por longitud 
#Dada una lista de palabras, ordénalas desde la más corta hasta la más larga. 
palabras = ['casa', 'elefante', 'sol', 'computadora', 'mesa', 'programacion'] 

ultima_palabra=[
    len(palabra)
for palabra in palabras
]
print(ultima_palabra)

listas_unidas=[]
for item in zip(palabras,ultima_palabra):
    listas_unidas.append(item)

sorted(listas_unidas, key=lambda x: x[1])
```

```python
#EJERCICIO 12 Eliminar elementos vacíos 
#Dada una lista con textos y strings vacíos, crea una nueva lista eliminando los elementos vacíos. 
textos = ['hola', '', 'python', '', 'listas', ' ', 'programar', ''] 
textos_limpiados=[]
for texto in textos:
    if texto.strip():
       textos_limpiados.append(texto)
    
print(textos_limpiados)
```

```python
#EJERCICIO 16. Dividir una lista en dos partes 
#Dada una lista de números, divídela en dos mitades. 
numeros = [10, 20, 30, 40, 50, 60, 70, 80] 
mitad_lista=round(len(numeros)/2)
print(mitad_lista)
primera_parte=[]
segunda_parte=[]

for idx, numero in enumerate(numeros):
    if idx<mitad_lista:
        primera_parte.append(numero)
print(primera_parte)

for idx, numero in enumerate(numeros):
    if idx>=mitad_lista:
        segunda_parte.append(numero)
print(segunda_parte)
```

```python
#EJERCICIO 17. Reemplazar elementos 
#Dada una lista de palabras, reemplaza todas las apariciones de 'java' por 'python'. 
palabras = ['java', 'python', 'java', 'c++', 'java', 'javascript'] 
lista_reemplazada=[]
for palabra in palabras:   
    lista_reemplazada.append(palabra.replace("java","python"))
print(lista_reemplazada)
```

```python
#EJERCICIO 19. Combinar dos listas alternando elementos 
#Dadas dos listas del mismo tamaño, crea una nueva lista alternando los elementos de ambas listas. 
numeros = [1, 2, 3, 4] 
letras = ['a', 'b', 'c', 'd'] 
resultado_esperado = [1, 'a', 2, 'b', 3, 'c', 4, 'd'] 
listas_intercalada=[]
for numero,letra in zip(numeros,letras):
    listas_intercalada.extend([numero,letra])
print(listas_intercalada)
```
